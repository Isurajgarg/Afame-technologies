```python
# distribution of attrition in company

hr_data['Attrition'].value_counts()
```

```python
# Percentage of attrition in company

np.round(hr_data['Attrition'].value_counts(normalize=True)*100,2)
```

Around 16% Employee leave the company while 83% Employee stay in the company



```python
numerical_columns = [col for col in hr_data.columns if hr_data[col].dtypes != "object" and col not in ['StandardHours', 'EmployeeCount']]
# Calculate the correlation matrix for numerical columns
plt.figure(figsize=(35,35))
sns.heatmap(hr_data[numerical_columns].corr(), annot=True)
```

```python
for col in ['EducationField','JobRole','Department']:
    counts = hr_data[col].value_counts()
    fig = plt.figure(figsize=(12,6))
    counts.plot(kind = 'bar', title=f'Distribution of {col}')
    plt.show()
```

```python
sns.histplot(hr_data['Age'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Attrition',hue='Department',data=hr_data)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Attrition',hue='Gender',data=hr_data)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Attrition',hue='MaritalStatus',data=hr_data)
plt.show()
```

```python
hr_data.groupby(['Attrition', 'Department']).size()
```

From this we can say the Employee from Research & Development Department leave the company most , then from Sales department and at last only few Employee from Human Resource leaves.
